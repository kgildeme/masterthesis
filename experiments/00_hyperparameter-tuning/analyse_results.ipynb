```python
import json

from cids.util import misc_funcs as misc
import matplotlib.pyplot as plt
import pandas as pd


pd.set_option('display.max_columns', 10)
```

```python
metric = "val/score"
mode = "max"
results = misc.load_optimization_results(os.path.join(misc.root(), "results/00_hyperparameter-tuning/OPTCFull/unsupervised/MLPAE/MLPAE-OPTC-1"), metric=metric, mode=mode)
# important_columns = ["epoch", "steps", "train/loss", "train/acc", "val/loss", "val/loss_agg", "val/loss_net", "val/best_loss", "val/acc", "val/acc_net", "lr", "model", "alpha"]
print(results.columns)
important_columns = ["epoch", "steps", "train/loss", "val/loss", "val/score", "val/best_loss", "lr", "model", "config"]
# important_columns = ["epoch", "steps", "train/loss", "val/loss", "val/score", "val/best_loss", "lr", "model"]
results = results[important_columns]
```

```python
print(4.5e-4)
```

```python
top_n = 100
top_results = results.nsmallest(top_n, metric) if mode == "min" else results.nlargest(top_n, metric)

display(top_results)
for i in range(5):
    print(json.dumps(top_results.iloc[i]["config"], indent=2))
```

```python

# Extract the hidden_dims from the model column
hp = 'hidden_dims'
# top_results[hp] = top_results['model'].apply(lambda x: str(x['encoder'][hp]))
top_results[hp] = top_results['model'].apply(lambda x: x[hp])
# Plot the corresponding values of lr for each unique hidden_dims
unique_hidden_dims = top_results[hp].unique()
for hidden_dim in unique_hidden_dims:
    subset = top_results[top_results[hp] == hidden_dim]
    plt.scatter(subset['lr'], subset[metric], label=f'{hp}: {hidden_dim}')

plt.xscale('log')
plt.xlabel('Learning Rate')
plt.ylabel(metric)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
