```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
x = np.linspace(-7.5, 7.5, 500)
y_sigmoid = 1 / (1 + np.exp(-x))

x_relu = np.linspace(-2, 5, 500)
y_relu = np.maximum(0, x_relu)
```

```python
# Create separate figures for each plot
fig1, ax1 = plt.subplots(figsize=(6, 4))
ax1.plot(x, y_sigmoid, color='#2E5A88', linewidth=2)
ax1.set_ylim(-0.1, 1.02)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.axhline(y=0, color='black', linestyle='-', alpha=0.3)
ax1.axvline(x=0, color='black', linestyle='-', alpha=0.3)
fig1.tight_layout()
fig1.savefig('plots/background/sigmoid_function.pdf')
plt.tight_layout()
plt.show()
plt.close(fig1)

fig2, ax2 = plt.subplots(figsize=(6, 4))
ax2.plot(x_relu, y_relu, color='#2E5A88', linewidth=2)
ax2.set_ylim(-0.5, 5.1)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('y', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)
ax2.axvline(x=0, color='black', linestyle='-', alpha=0.3)
fig2.tight_layout()
fig2.savefig('plots/background/relu_function.pdf')
plt.tight_layout()
plt.show()
plt.close(fig1)
```

```python
x_mse = np.linspace(-5, 5, 200)
y_mse = x_mse ** 2
y_inverse_mse =-x_mse**2
y_reciprocal_mse = 1 / (x_mse**2 + 1e-12)
```

```python


# Plot inverse MSE with MSE in the background
fig4, ax4 = plt.subplots(figsize=(6, 4))
ax4.plot(x_mse, y_mse, color='gray', linewidth=1, linestyle='--', label='MSE')
ax4.plot(x_mse, y_inverse_mse, color='#2E5A88', linewidth=2, label='Inverse MSE')
ax4.set_ylim(-4.5, 4.5)
ax4.set_xlabel('x', fontsize=12)
ax4.set_ylabel('y', fontsize=12)
ax4.grid(True, linestyle='--', alpha=0.7)
ax4.axhline(y=0, color='black', linestyle='-', alpha=0.3)
ax4.axvline(x=0, color='black', linestyle='-', alpha=0.3)
ax4.legend(loc='upper right')
fig4.tight_layout()
fig4.savefig('plots/background/inverse_mse_with_mse.pdf')
plt.tight_layout()
plt.show()
plt.close(fig4)

# Plot reciprocal MSE with MSE in the background
fig5, ax5 = plt.subplots(figsize=(6, 4))
ax5.plot(x_mse, y_mse, color='gray', linewidth=1, linestyle='--', label='MSE')
ax5.plot(x_mse, y_reciprocal_mse, color='#2E5A88', linewidth=2, label='Reciprocal MSE')
ax5.set_ylim(-0.5, 10.5)  # Adjusted scale for reciprocal
ax5.set_xlabel('x', fontsize=12)
ax5.set_ylabel('y', fontsize=12)
ax5.grid(True, linestyle='--', alpha=0.7)
ax5.axhline(y=0, color='black', linestyle='-', alpha=0.3)
ax5.axvline(x=0, color='black', linestyle='-', alpha=0.3)
ax5.legend(loc='upper right')
fig5.tight_layout()
fig5.savefig('plots/background/reciprocal_mse_with_mse.pdf')
plt.tight_layout()
plt.show()
plt.close(fig5)
```
